```python
%matplotlib inline
import requests as re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
import time
from bs4 import BeautifulSoup as bs
url ='https://www.8591.com.tw/mallList-list.html?&group=1&searchType=0&priceSort=0&ratios=0&searchGame=859&searchServer=0&firstRow=0&totalRows=459'

header ={
    'User-Agent':
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'
    }
def get_index(url):
    res =re.get(url)
    sp = bs(res.text,'html.parser')
    sp.encoding = 'utf-8'
    tle = sp.select('a')

    df = []
    for tts in tle:
        titles = tts.get('title')
        if titles == None:
            not print
        else:
#        print(titles)
            df.append(titles)
    return df

def get_soup(url):
    soups =[]
#    b = self.get_index()
    try:
        for page in range(0,5):
            html = re.get('https://www.8591.com.tw/mallList-list.html?&group=1&searchType=0&priceSort=0&ratios=0&searchGame=859&searchServer=0&firstRow=' + str(page) + '&totalRows=459',
                         headers = header)
            if html.status_code == 200:
                sps = bs(html.text,'lxml')
                sps.encoding= 'utf-8'
                tit = sps.find_all('a')
                for ttss in tit:
                    titless = ttss.get('title')
                    if titless == None:
                        not print
                    else:
                        soups.append(titless)
                        time.sleep(2)
            else:
                print('網站無法讀取！')
#                html.raise_for_status()
        return soups
    except Exception as e:
        print(e)
        sys.exit()
```
