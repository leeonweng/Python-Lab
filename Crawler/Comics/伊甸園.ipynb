```python
def get_tophref(url):
    dff = []
    try:
        r = req.get(url, headers = headers)
        r.encoding = 'utf-8'
        s = bs(r.text, 'lxml')
        d = s.select('a')
#print(d)

        for gg in d:
            a = gg.get('href')
#        print(a)
            if ('/comic/checkview') in a:
                n = re.sub(r'\D', '', a)
                b = dff.append(n[:-1])
            else:
                not print
        return dff
    except Exception as e:
        print(e)
        sys.exit()
```
