```python
x = [1, 5, 9, 16]
y1 = [-8, 2 ,6 ,10]
y2 = [-3, 4 ,9 ,13]
```

```python
import numpy as np

plt.figure(figsize=(10,5))
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')

plt.legend(loc = 'best')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('titlesss')
plt.xlim(-2,20)
plt.ylim(-10,20)
plt.xticks(np.arange(-2,20,2))
plt.yticks(np.arange(-10,20,2))

plt.show()
```
