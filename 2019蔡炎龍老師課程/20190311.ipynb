```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact_manual
```

```python
X = np.random.rand(6)
Y = np.random.rand(6)
```

```python
plt.scatter(X, Y, s=100 , c='r');
#s= 點的改變大小 c= 改變顏色
```

```python
x =  np.linspace(0, 1, 1000)
```

```python
def draw(n):
    plt.scatter(X, Y, s=100 , c='r')
    y = 0.5*np.sin(n*x*2*np.pi) + 0.5
    plt.plot(x, y)
    plt.show()
```

```python
interact_manual(draw, n =(1., 100.))
```

# BMI計算器

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact_manual
```

```python
def BMI_cal(身高, 體重):
    身高 = float(身高)/100
    體重 = float(體重)
    bmi = 體重 / (身高**2)
    print("你的 BMI 是 = {:.2f}".format(bmi))
```

```python
interact_manual(BMI_cal, 身高 = '請輸入你的身高:(cm)', 體重 = '請輸入你的體重:(kg)')
```
